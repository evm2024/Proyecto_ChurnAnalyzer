```python
import pandas as pd
import os

# Ruta al directorio de datos compartido entre Airflow y Jupyter
DATA_DIR = '/home/jovyan/work/data'

# Archivo Excel
excel_file = os.path.join(DATA_DIR, 'Telco_customer_churn.xlsx')

if os.path.exists(excel_file):
    
    # Cargar dataset desde Excel
    df = pd.read_excel(excel_file)

    print(f"Dataset cargado: {len(df)} filas y {len(df.columns)} columnas")

    # Mostrar primeras filas
    display(df.head())

    # Información general del dataset
    print("\nInformación del dataset:")
    df.info()

    # Mostrar nombres de columnas
    print("\nColumnas del dataset:")
    print(df.columns)

else:
    print(f"Archivo no encontrado en: {excel_file}")
    print(f"\nArchivos disponibles en {DATA_DIR}:")
    
    if os.path.exists(DATA_DIR):
        for file in os.listdir(DATA_DIR):
            print(f"  - {file}")
    else:
        print(f"El directorio {DATA_DIR} no existe")
```

```python
# Copia del dataset para limpieza
df_clean = df.copy()
```

```python
# 1. Quitar espacios en nombres de columnas
df_clean.columns = df_clean.columns.str.replace(" ", "_")
```

```python
# 2. Convertir Total_Charges a numérico
df_clean["Total_Charges"] = pd.to_numeric(df_clean["Total_Charges"], errors="coerce")
```

```python
# 3. Revisar valores nulos
print("Valores nulos por columna:")
print(df_clean.isnull().sum())
```

```python
# 3. Revisar valores nulos
print("Valores nulos por columna:")
print(df_clean.isnull().sum())
```

```python
df_clean["Churn_Reason"] = df_clean["Churn_Reason"].fillna("No Churn")
```

```python
print(df_clean.columns.tolist())
```

```python
df_clean = df_clean.dropna(subset=["Total_Charges"])
```

```python
# 3. Revisar valores nulos
print("Valores nulos por columna:")
print(df_clean.isnull().sum())
```

```python
cols_drop = [
    "Count",
    "Lat_Long",
    "Zip_Code",
    "Country",
    "State",
    "City",
    "Churn_Reason"
]

df_clean = df_clean.drop(columns=cols_drop)
```

```python
df_clean = df_clean.drop_duplicates()
```
